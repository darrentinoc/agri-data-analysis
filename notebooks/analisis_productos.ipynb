# Análisis Exploratorio de Datos Agrícolas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

"""
Leer DataSet
"""
df = pd.read_csv('/workspaces/agri-data-analysis/data/productos_agricolas.csv', parse_dates =['fecha_cosecha'])
df.head()
```

## Estadísticas decriptivas

```python
df.describe()
```

## Distribución de Fechas de Cosecha

```python
df ['fecha_cosecha'].hist(bins=30, figsize=(10,5))
plt.title('Distribución de Fechas de Cosecha')
plt.ylabel('Frecuencia')
plt.show()
```

## Rendimiento promedio por Región

```python
region_mean = df.groupby('region')['rendimiento_kg_ha'].mean()
region_mean.plot(kind='bar', figsize=(8,5))
plt.title('Rendimiento Promedio por Región(kg/ha)')
plt.ylabel('kg/ha')
plt.show()
```

## Precio Promedio por Tipo de Producto

```python
tipo_precio = df.groupby('tipo')['precio_kg'].mean()
tipo_precio.plot(kind='barh',figsize=(8,5))
plt.title('Precio Promedio por Tipo de Producto (S/ por kg)')
plt.xlabel('S/ por kg')
plt.show()
```

## Correlación entre Rendimiento y Precio

```python
sns.scatterplot(data=df, x='rendimiento_kg_ha', y='precio_kg', hue='region')
plt.title('Rendimiento vs Precio por Región')
plt.show()
```
